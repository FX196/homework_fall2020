```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_section_results(file, keys=['Eval_AverageReturn']):
    eval_returns_dict = {}
    for k in keys:
        eval_returns_dict[k] = []
    for e in tf.train.summary_iterator(file):
        for v in e.summary.value:
            if v.tag in keys:
                eval_returns_dict[v.tag].append(v.simple_value)
    for k in keys:
        eval_returns_dict[k] = np.array(eval_returns_dict[k])
    return eval_returns_dict
```

```python
q1_sb_files = {"small_scale_data/q1_sb_no_rtg_dsa_CartPole-v0_26-09-2020_18-36-19/events.out.tfevents.1601170579.heisenberg": "no_rtg_dsa", 
               "small_scale_data/q1_sb_rtg_dsa_CartPole-v0_26-09-2020_18-36-17/events.out.tfevents.1601170577.heisenberg": "rtg_dsa", 
               "small_scale_data/q1_sb_rtg_na_CartPole-v0_26-09-2020_18-36-13/events.out.tfevents.1601170573.heisenberg": "rtg_na"}

q1_lb_files = {"small_scale_data/q1_lb_no_rtg_dsa_5000_CartPole-v0_26-09-2020_18-36-15/events.out.tfevents.1601170575.heisenberg": "no_rtg_dsa", 
               "small_scale_data/q1_lb_rtg_dsa_5000_CartPole-v0_26-09-2020_18-36-09/events.out.tfevents.1601170569.heisenberg": "rtg_dsa",
               "small_scale_data/q1_lb_rtg_na_5000_CartPole-v0_26-09-2020_18-36-08/events.out.tfevents.1601170568.heisenberg": "rtg_na"}
```

```python
def plot_q5(files_label_dict, title):
    for f in files_label_dict.keys():
        returns_dict = get_section_results(f, keys=['Eval_AverageReturn', 'Eval_StdReturn'])
        eval_returns, stds = returns_dict['Eval_AverageReturn'], returns_dict['Eval_StdReturn']
        plt.plot(range(len(eval_returns)), eval_returns, label=files_label_dict[f])
    
    plt.xlabel("Training Steps")
    plt.ylabel("Eval_AverageReturn")
    plt.title(title)
    plt.legend()
    
    plt.show()
    
```

```python
plot_q5(q1_sb_files, title="Small Batch Evaluation Returns")
plot_q5(q1_lb_files, title="Large Batch Evaluation Returns")
```
